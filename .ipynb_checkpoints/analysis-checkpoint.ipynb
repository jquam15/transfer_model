**The first thing I'm going to do is actually find the effective transfers. I'm going to evaluate someone as an "effective transfer" by (1) finding the players that actually played consistent minutes (already know they played 75% of their team's games) and (2) were additive when they played (positive BPM players). I'm going to set the MPG cutoff at 14 minutes per game as that is ~1/3 of a game. This is obviously arbitrary, and the threshold could be increased to find something like "stars" as opposed to just effective rotation players in general (stars included). This is mostly a curiosity thing.**

```python
#take the necessary columns
effective = df[["Name", "Position", "Class", "Previous School", "New School", "Season", "Conf", "MPG", "OBPM", 
                "DBPM", "BPM"]]
#only looking at the season after players transferred (post-transfer season)
effective = effective[effective.Season == "2021-22"]
#filter
effective = effective[(effective.MPG >= 14) & (effective.BPM > 0)]
#I don't have a great way of dealing with players who share the same name so I'm going to drop them
#There ar only like 6 repeat players, so it doesn't impact sample size in a major way
repeat_names = list(effective[effective.duplicated("Name") == True]["Name"])
effective = effective[~effective["Name"].isin(repeat_names)]
#do a quick check to make sure there are no left over repeated names
assert len(effective[effective.duplicated("Name", keep=False) == True]) == 0
effective = effective.sort_values(by="BPM", ascending=False)
effective.head(10)
```

**I'm going to add two variables. The first will be "level" which is the level that they are transferring from. Level will be "high" for Power 6 schools, "mid" for mid-majors, and "low" for low-majors. Low majors are schools that are 1 bid leagues while mid-majors are non Power-5 schools that usually receive more than 1 bid. The second will be "move" which will either be "same" (high --> high), "up" (mid --> high), "2 up" (low --> high), "down" (high --> mid), or "2 down" (high --> low).**

```python
#this funtion takes the value of the "Post_Conf" column for a player and gives the conf a competition categorization
#of "high", "mid", or "low"
def define_level(conf):
    if conf in ["ACC", "SEC", "Pac-12", "Big 12", "Big Ten", "Big East"]:
        return("high")
    if conf in ["AAC", "WCC", "MWC", "A-10"]:
        return("mid")
    else:
        return("low")

#take only the pre-transfer year since the goal is to predict post-transfer success from pre-transfer data
transfers = transfers[transfers.Season == "2020-21"]
transfers = transfers.reset_index(drop=True)
#make pre_level and post_level column for each player by using apply to apply the "define_level" function defined above
transfers["pre_level"] = transfers["Conf"].apply(define_level)
transfers["post_level"] = transfers["Post_Conf"].apply(define_level)
#initialize empty column called "move" that will categorize how the level of competition changed when they transferred
transfers["move"] = ""

#loop through each player
for i in range(len(transfers)):
    #grab their pre_level and post_level statuses 
    pre_level = transfers.iloc[i]["pre_level"]
    post_level = transfers.iloc[i]["post_level"]
    #if they moved to the same level give them a "same" categorization
    if pre_level == post_level:
        transfers.at[i, "move"] = "same"
    #else if they moved up 1 level give them "up"
    elif (pre_level == "mid" and post_level == "high") or (pre_level == "low" and post_level == "mid"):
        transfers.at[i, "move"] = "up"
    #else if they moved down 1 level give them "down"
    elif (pre_level == "high" and post_level == "mid") or (pre_level == "mid" and post_level == "low"):
        transfers.at[i, "move"] = "down"
    #else if they went up two levels, give them "2 up"
    elif pre_level == "low" and post_level == "high":
        transfers.at[i, "move"] = "2 up"
    #if non of those conditional statements has hit, then we know they had to have moved down 2 levels --> "2 down" categorization
    else:
        transfers.at[i, "move"] = "2 down"
#inspect the first couple rows to ensure things went they way I wanted them to
transfers.head()
```

**This code does the same thing as the code above, but for transfers2 (2021-22 pre and 2022-23 post transfers)**

```python
#take only the pre-transfer year since the goal is to predict post-transfer success from pre-transfer data
transfers2 = transfers2[transfers2.Season == "2021-22"]
transfers2 = transfers2.reset_index(drop=True)
#make level and move column
transfers2["pre_level"] = transfers2["Conf"].apply(define_level)
transfers2["post_level"] = transfers2["Post_Conf"].apply(define_level)
transfers2["move"] = ""

for i in range(len(transfers2)):
    pre_level = transfers2.iloc[i]["pre_level"]
    post_level = transfers2.iloc[i]["post_level"]
    if pre_level == post_level:
        transfers2.at[i, "move"] = "same"
    elif (pre_level == "mid" and post_level == "high") or (pre_level == "low" and post_level == "mid"):
        transfers2.at[i, "move"] = "up"
    elif (pre_level == "high" and post_level == "mid") or (pre_level == "mid" and post_level == "low"):
        transfers2.at[i, "move"] = "down"
    elif pre_level == "low" and post_level == "high":
        transfers2.at[i, "move"] = "2 up"
    else:
        transfers2.at[i, "move"] = "2 down"
    
transfers2.head()
```

# Modeling Starts Here

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, ElasticNet, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.svm import SVR
```

**I'm going to start with feature importances to get a guiding sense of what explanatory variables to look into further**

```python
#drop all identifying columns since we only want the actual features
features = transfers.drop(columns=["Position", "Name", "Class", "Previous School", "New School"])
#Split into X and y data frames for modeling purposes
features_X = features.loc[:, features.columns != "Post_BPM"]
features_y = features["Post_BPM"]
#inspect the features
features_X.head()
```

### Encoding Categorical Variables

**Since there are some categorical variables that I want to use to predict, I need to encode these variables as the regression cannot understand variables that are not numeric. Therefore, I create a column for each individual category of "move" and "level". I encode the column corresponding to an individual category with the value of 1 if the corresponding player falls under that category, and 0 if the category is not the case for the corresponding player. OneHotEncoder does that process for me, and then I add these columns to the data**

```python
#set up the encoder
enc = OneHotEncoder()
#fit and transform the level column (creates a dataframe with 3 columns-1 for each category of the "Level" column)
enc_level = pd.DataFrame(enc.fit_transform(features_X[["Level"]]).toarray())
#rename columns, encoded variables in alphabetical order (high is 0, low is 1, mid is 2)
enc_level = enc_level.rename(columns={0: "high", 1: "low", 2: "mid"})
#fit and transform the "Move" columns (creates a dataframe with 5 columns-1 for each category of the "Move" column)
enc_move = pd.DataFrame(enc.fit_transform(features_X[["Move"]]).toarray())
#for the move case (2 down is 0, 2 up is 1, down is 2, same is 3, up is 4)
enc_move = enc_move.rename(columns={0: "2 down", 1: "2 up", 2: "down", 3: "same", 4: "up"})
#add the 8 total encoded columns to the dataframe with the rest of the features
features_X = features_X.join(enc_level).join(enc_move)
#since we encoded the "Level" and "Move" columns, we don't need the originals anymore as their data is contained 
#in the format I want it in now
features_X = features_X.drop(columns=["Level", "Move"])
#print all the features
print(features_X.columns)
#inspect the first couple of rows
features_X.head()
```

```python
#initialize the random forest object and give it a random state for reproducibility purposes
rf = RandomForestRegressor(random_state=42)
#fit the random forest
rf.fit(features_X, features_y)
#get the values of the feature importances from the fitted random forest model
feat_importances = pd.Series(rf.feature_importances_, index=features_X.columns).nsmallest(30)
#size the plot
fig, ax = plt.subplots(figsize=(6, 5))
#plot the feature importances as a horizontal bar plot
feat_importances.plot(kind="barh", ax=ax)
#set the title and axist lables
ax.set_title("Feature Importances using MDI")
ax.set_xlabel("Mean decrease in impurity")
None
```

```python
#create a list of the features that I want to use to predict the post_transfer_BPM
prediction_columns = ["MPG", "BPM", "PProd", "Level", "Move", "TS%"]
#create an X dataframe with the prediction columns
X = transfers[prediction_columns]
#create the array of target data
y = transfers["Post_BPM"]
```

**Need to use one hot encoding to encode categorical variables "Level" and "Move". This is the same process used earlier when I created the feature importances plot, but now I'm encoding the dataframes I'm going to use for prediction**

```python
enc = OneHotEncoder()
if ("Move" in prediction_columns) and ("Level" in prediction_columns):
    enc_level = pd.DataFrame(enc.fit_transform(X[["Level"]]).toarray())
    #rename columns, encoded variables in alphabetical order (high is 0, low is 1, mid is 2)
    enc_level = enc_level.rename(columns={0: "high", 1: "low", 2: "mid"})
    enc_move = pd.DataFrame(enc.fit_transform(X[["Move"]]).toarray())
    #for the move case (2 down is 0, 2 up is 1, down is 2, same is 3, up is 4)
    enc_move = enc_move.rename(columns={0: "2 down", 1: "2 up", 2: "down", 3: "same", 4: "up"})
    X = X.join(enc_level).join(enc_move)
    X = X.drop(columns=["Level", "Move"])
    
elif "Level" in prediction_columns:
    enc_level = pd.DataFrame(enc.fit_transform(X[["Level"]]).toarray())
    #rename columns, encoded variables in alphabetical order (high is 0, low is 1, mid is 2)
    enc_level = enc_level.rename(columns={0: "high", 1: "low", 2: "mid"})
    X = X.join(enc_level)
    X = X.drop(columns="Level")
    
elif  "Move" in prediction_columns:
    enc_move = pd.DataFrame(enc.fit_transform(X[["Move"]]).toarray())
    #for the move case (2 down is 0, 2 up is 1, down is 2, same is 3, up is 4)
    enc_move = enc_move.rename(columns={0: "2 down", 1: "2 up", 2: "down", 3: "same", 4: "up"})
    X = X.join(enc_move)
    X = X.drop(columns="Move")
    
X.head()
```

### Scaling the Data

**There is a SettingWithCopyWarning when I try to scale the data, so the 2 cells below check to ensure that I'm scaling as intended. I'm taking the first 5 rows and then manually standardizing them and checking to make sure the standardized columns in X match up with the manually calculated ones. Each ROW (array) of the manual output corresponds to the COLUMN output of X. First printed array output (row) is the MPG column.**

```python
X_train[0:4]
```

**One possible point of confusion for the manual calculation is that StandardScaler uses the ddof=0 parameter when calculating the standard deviation. If you would just use .std() with no parameter you would get slightly different values (only by ~.001). This small of a difference causes no major impact in this case. I added the parameter so the manually calculated standardization matches up exactly with the StandardScaler standardization.**

```python
#perform manual standardization by taking the (values - mean) / standard deviation (first four values for each column)
print(([31.5, 35.8, 36, 35.1] - X_train["MPG"].mean()) / X_train["MPG"].std(ddof=0))
print(([-.8, 7.1, -1.6, 5.3] - X_train["BPM"].mean()) / X_train["BPM"].std(ddof=0))
print(([179, 556, 284, 443] - X_train["PProd"].mean()) / X_train["PProd"].std(ddof=0))
print(([.609, .517, .593, .538] - X_train["TS%"].mean()) / X_train["TS%"].std(ddof=0))
```

**Only scaling the numerical data. This scaling is important specifically with regression. The variables that I encoded with 1s and 0s do not need to be scaled. So, move and level are not scaled, hence they do not appear is the list object "cols_to_scale"**

```python
%%capture --no-display 
#^^^double checked the warning and everything went as intended so I'm suppressing the warning message

#grab columns as a list
cols_to_scale = ["MPG", "BPM", "PProd", "TS%"]

#create and fit scaler
scaler = StandardScaler()
scaler.fit(X_train[cols_to_scale])

#scale selected data according to the training fit so that the test set remains "unseen"
X_train.loc[:, cols_to_scale] = scaler.transform(X_train[cols_to_scale])

#transforming based on the fit of the training data
X_test.loc[:, cols_to_scale] = scaler.transform(X_test[cols_to_scale])

X_train.head()
```

## Models

**The metrics that I chose to use to evaluate the models were R^2 (Coefficient of Determination) and MAE (Mean Absolute Error). I picked MAE as opposed to MSE (mean squared error) or RMSE (root mean squared error) because there are a decent amount of outliers (some players drastically improve their skills from year to year which increases performance for example) and MAE puts less emphasis on outliers (they aren't squared like they are with MSE).**

```python
#initialize linear regression object
lr = LinearRegression()
#fit the training data
lr.fit(X_train, y_train)
#use the training data to predict on the test data
y_pred = lr.predict(X_test)
#calculate performance metrics (R^2: coefficient of determination and MAE: mean absolute error)
r_squared = lr.score(X_test, y_test)
mse = mean_absolute_error(y_true = y_test, y_pred = y_pred)
#this prints the metrics
print("The value of R^2 is " + str(r_squared))
print("The MAE is " + str(mse))
```

**I wanted to compare Ridge regression to the previous linear regression output because some of my features are decently correlated to each other. Ridge regression is more robust in cases where features are highly correlated. By looking at the output though, there does not appear to be much of a difference in the scores of linear and ridge regression.**

```python
#initialzie the ridge object
ridge = Ridge()
#fit on the training
ridge.fit(X_train, y_train)
#use the training data to evaluate on the test data
y_pred_r = ridge.predict(X_test)
#get metrics
r_squared_r = ridge.score(X_test, y_test)
mse_r = mean_absolute_error(y_true = y_test, y_pred = y_pred_r)
#print the metrics
print("The value of R^2 is " + str(r_squared_r))
print("The MAE is " + str(mse_r))
```

```python
#The goal of this project is to predict "positive impact transfers". So this function takes a BPM value passed as 
#the "target" parameter and then returns an accuracy score. The accuracy score is the number of transfers my model 
#predicted to be at the "target" threshold or above and actually were divided by the total of predicted positive transfers
def predicted_positive_accuracy(target_BPM):
    #take the number of players the model predicted to be positive
    predicted_pos = len(pred[pred.Predicted >= target_BPM])
    #get the number of players where my model predicted them to be positive but they actually weren't
    misses = len(pred[(pred.Predicted >= target_BPM) & (pred.Post_BPM < target_BPM)]) 
    #return an accuracy score
    return(1 - (misses/predicted_pos))

predicted_positive_accuracy(1.5)
```

**Compare to a naive model where we just use the prior BPM being equal to or over the target to determine if the next season will also be equal to or over the target**

```python
naive_target = 1.5
naive_predicted_pos = len(features[features.Post_BPM >= naive_target]) 
#number of times where just using previous years BPM to predict next season BPM failed
naive_misses = len(features[(features.BPM >= naive_target) & (features.Post_BPM < naive_target)])
1 - (naive_misses / naive_predicted_pos)
```
